```python
import torch 
import torch.nn as nn
```

```python
X_train = torch.tensor(X_train.values, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.long)
X_test = torch.tensor(X_test.values, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.long)
```

```python
class IPLNet(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, output_dim=None, dropout=0.3):
        super().__init__()
        if output_dim is None:
            output_dim = len(set(y))

        # Hidden layers
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.fc3 = nn.Linear(hidden_dim // 2, hidden_dim // 4)
        self.fc4 = nn.Linear(hidden_dim // 4, hidden_dim // 8)
        self.fc_out = nn.Linear(hidden_dim // 8, output_dim)

        # Activations + dropout
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        # x = self.dropout(x)  
        x = self.relu(self.fc2(x))
        # x = self.dropout(x)
        x = self.relu(self.fc3(x))
        # x = self.dropout(x)
        x = self.relu(self.fc4(x))
        x = self.fc_out(x)
        return x
```

```python
def init_weights(m, mode = 'xavier'):
    if isinstance(m, nn.Linear):
        if mode == "xavier":
            nn.init.xavier_uniform_(m.weight)
        elif mode == "kaiming":
            nn.init.kaiming_normal_(m.weight, nonlinearity="relu")
        elif mode == "normal":
            nn.init.normal_(m.weight, mean=0, std=0.01)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0)
```

```python
X_train.shape
```

```python
import torch.optim as optim 
```

```python
def train_model(init_mode):
    model = IPLNet(input_dim = X_train.shape[1])
    model.apply(lambda m: init_weights(m, mode=init_mode))
    
    loss_fn = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    epochs = 130
    history = {"loss": [], "acc":[]}
    best_preds = None
    
    for epoch in range(epochs):
        model.train()
        optimizer.zero_grad()
        outputs = model(X_train)
        loss = loss_fn(outputs, y_train)
        loss.backward() 
        optimizer.step()
        
        model.eval()
        with torch.no_grad():
            preds = torch.argmax(model(X_test), dim=1)
            acc = (preds == y_test).float().mean().item()
            
        
        history["loss"].append(loss.item())
        history['acc'].append(acc)
        
        if(epoch+1) % 5== 0:
            print(f"[{init_mode}] Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f},Test Acc: {acc:.4f}")
        
        if epoch == epochs-1:
            best_preds = preds.cpu().numpy()
            
    return history , best_preds
```

```python
results = {}
predictions = {}
for mode in ["xavier", "kaiming", "normal", "uniform"]:
    results[mode], predictions[mode] = train_model(mode)
```
