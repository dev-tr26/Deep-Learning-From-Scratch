- visualising convolution - simple edge filter

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img_path = r"C:\Desktop\images\WhatsApp Image 2024-12-16 at 13.10.14_c1327737.jpg"

img = mpimg.imread(img_path)

# decrease dim of img (more suitable for filter)
img = img[::7, ::7, :]

# show pixel values r g b for 5 pixels
print(img[0, :5])
```

```python
plt.imshow(img)
```

```python
import numpy as np
import copy

filter = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
    ])

H,W, C = img.shape

filtered_img = np.zeros((H-2, W-2), dtype=int)  # output only for first channel


for row in range(H-2):  
    for col in range(W-2):   
        
        # extract 3x3 patch from 1st channel 
        patch = img[row:row+3, col:col+3, 0]
        
        # apply filter
        conv_result = np.sum(patch*filter)
        
        filtered_img[row,col]  = conv_result
```

```python
plt.imshow(filtered_img, vmin=filtered_img.min(), vmax=filtered_img.max() )
```
